### Q1. Account 클래스 : 은행에 가서 계좌를 개설하면 은행 이름, 예즘주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후, 생성자를 구현해 보세요. 생성자에는 초기 잔액만 입력 받습니다. 은행 이름은 SC은행으로 계좌번호는 3자리-3자리-6자리 형태로 랜덤하게 생성됩니다. (은행이름: SC은행, 계좌번호: 111-11-111111)

  * name과 bal 매개변수 사용
* 계좌번호는 랜덤 값으로 생성하되 각 자리수가 '-'로 구분되어야 한다.

### Q4. 입금 메서드 : Account 클래스에 입금을 위한 deposit 메서드를 추가하세요. 입금인 최소 1원만 가능합니다.


#### 고려했던 사항
* 입금 금액이 0이면 다시 입력을 받아야 한다.
  * while 반복문으로 입금 금액이 0인 경우 재입력 유도
* 입금 금액은 정수로 받아야 한다.
  * try - except 구문으로 문자 입력 시, 재입력 유도
* 입금금액은 계좌 잔액에 추가되어야 한다.
  * 계좌 잔액 += 입금 금액

---
```
입금할 금액을 입력하세요: 입금 금액 입력은 숫자만 가능합니다.
처리가 완료되었습니다.
이선재님의 잔고는 1.0원 입니다.
```

````python
# 예금 입금 메서드
def deposit(self):
    # 초기 예금 값 0으로 초기화
    self.dps_money = 0

    try:
        # 입력한 예금이 0이 아닐 때까지 반복
        while self.dps_money == 0:
            # 입력한 예금을 정수형으로 캐스팅
            self.dps_money = int(input('입금할 금액을 입력하세요: '))

            # 입력한 예금이 0일 경우 재입력 유도
            if self.dps_money == 0:
                print('계좌 입금은 1원 이상부터 가능합니다.')
            else:
                # 예금을 잔고에 추가
                self.bal += self.dps_money

            # 처리 결과 출력
        print(f'''처리가 완료되었습니다.
{self.name}님의 잔고는 {format(self.bal, ',')}원 입니다.''')

        return True
    except:
        # 문자형이 입력되었을 때 안내 출력
        print('입금 금액 입력은 숫자만 가능합니다.')

        return False



# 계좌 입금 함수 정의
def deposit(client):
    # 정상 입력 여부 boolean 초기화
    dps_success = False
    # 처리 결과로 True가 반환될 때까지 반복
    while dps_success == False:
        if client.deposit():
            dps_success = True

            break
        else:
            continue

client_a = Account('이선재', 0)
deposit(client_a)
````

### Q5. 출금 메서드 : Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요, 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.

#### 고려했던 사항
* 계좌 잔고 이상의 출금을 막는 조건문을 만들어야 한다.
  * 객체의 잔고 속성과 비교하여 조건문 생성
* 예금 출금은 정수형으로 받아야 한다.
  * try - except 구문으로 정수 입력 유도
* 출금된 금액만큼 잔고에서 차감되어야 한다.
  * 계좌 잔액 -= 출금액

---
```
출금할 금액을 입력하세요: d
입금 금액 입력은 숫자만 가능합니다.
출금할 금액을 입력하세요: 1
처리가 완료되었습니다.
이선재님의 잔고는 5.05원 입니다.
```

````python
# 출금 메서드 정의
def withdraw(self):

    # 출금 금액을 잔고보다 1 크게 초기화
    self.wd_money = self.bal + 1

    try:
        # 출금액이 잔고보다 작을 때까지 반복
        while self.wd_money > self.bal:
            self.wd_money = int(input('출금할 금액을 입력하세요: '))

            # 출금액이 잔고보다 클 경우, 재입력 유도
            if self.wd_money > self.bal:
                print(f'''계좌 출금은 잔액을 상회할 수 없습니다.
현재 {self.name}님의 잔고는 {format(self.bal, ',')}원 입니다.''')
            else:
                # 잔고에서 출금액 차감
                self.bal -= self.wd_money

        # 처리 결과 안내
        print(f'''처리가 완료되었습니다.
{self.name}님의 잔고는 {format(self.bal, ',')}원 입니다.''')

        return True
    except:
        # 문자 입력 시, 정수 입력 유도
        print('입금 금액 입력은 숫자만 가능합니다.')

        return False


# 계좌 출금 함수 정의
def withdraw(client):
    # 정상 입력 여부 boolean 초기화
    wd_success = False
    # 처리 결과가 True일 때까지 반복
    while wd_success == False:
        if client.withdraw():
            wd_success = True

            break
        else:
            continue

client_a = Account('이선재', 100000)
deposit(client_a)
````

### Q7. 이자 지급하기. 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.

#### 고려했던 사항
* 입금 횟수가 5의 배수가 될 때마다 이자를 지급해야 한다.
  * 입금 카운드 변수를 생성하여 입금이 완료된 시점에 1씩 증가 시킴
  * 입금 카운크가 5의 배수일 경우 이자 계산
* 이자율을 계산하는 연산식을 사용해야 한다.
  * 이자 = 원금 * 0.01
  * 잔고 += 이자


---
```
입금할 금액을 입력하세요: 100
처리가 완료되었습니다.
이선재님의 잔고는 100원 입니다.
입금할 금액을 입력하세요: 100
처리가 완료되었습니다.
이선재님의 잔고는 200원 입니다.
입금할 금액을 입력하세요: 100
처리가 완료되었습니다.
이선재님의 잔고는 300원 입니다.
입금할 금액을 입력하세요: 100
처리가 완료되었습니다.
이선재님의 잔고는 400원 입니다.
입금할 금액을 입력하세요: 100
처리가 완료되었습니다.
이선재님의 잔고는 505.0원 입니다.
```

````python
# 예금 입금 메서드
def deposit(self):
    # 초기 예금 값 0으로 초기화
    self.dps_money = 0

    try:
        # 입력한 예금이 0이 아닐 때까지 반복
        while self.dps_money == 0:
            # 입력한 예금을 정수형으로 캐스팅
            self.dps_money = int(input('입금할 금액을 입력하세요: '))

            # 입력한 예금이 0일 경우 재입력 유도
            if self.dps_money == 0:
                print('계좌 입금은 1원 이상부터 가능합니다.')
            else:
                # 예금을 잔고에 추가
                self.bal += self.dps_money
                # 입금이 완료된 시점에 입금 횟수 카운트
                self.dps_cnt += 1

        # 입금 카운드가 5의 배수일 경우 이자 계산
        if self.dps_cnt % 5 == 0:
          # 이자 = 원금의 1%
          interest = self.bal * 0.01
          # 원금에 이자 추가
          self.bal += interest

        # 처리 결과 출력
        print(f'''처리가 완료되었습니다.
{self.name}님의 잔고는 {format(self.bal, ',')}원 입니다.''')

        return True
    except:
        # 문자형이 입력되었을 때 안내 출력
        print('입금 금액 입력은 숫자만 가능합니다.')

        return False



# 계좌 입금 함수 정의
def deposit(client):
    # 정상 입력 여부 boolean 초기화
    dps_success = False
    # 처리 결과로 True가 반환될 때까지 반복
    while dps_success == False:
        if client.deposit():
            dps_success = True

            break
        else:
            continue

client_a = Account('이선재', 0)
deposit(client_a)
````
